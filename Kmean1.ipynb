```python
import pandas as pd
import math as mt
# import numpy
data = pd.read_csv('kmean.csv')
print(data)
K=3
nc_x=[6.2,6.6,6.5]
nc_y=[3.2,3.7,3.0]
c_x=[]
c_y=[]
sub1=1
sub2=1
k=1
while(sub1!=0 and sub2!=0 ):
    print("Iteration :",k)
    arr=[]
    c_x=nc_x
    c_y=nc_y

    for i in range (len(data)):
        dst=[]
    
        j=0
        while(j<3):
            dis=mt.sqrt((data['x1'][i]-c_x[j])**2+(data['x2'][i]-c_y[j])**2)
            dst.append(dis)
            j+=1
        m=dst.index(min(dst))
        arr.append(m+1)
    print(arr)
  

    c1=arr.count(1)
    c2=arr.count(2)
    c3=arr.count(3)
    sx1=0
    sy1=0
    sx2=0
    sy2=0
    sx3=0
    sy3=0
    nc_x=[]
    nc_y=[]
    for i in range(len(arr)):
        if(arr[i]==1):
            sx1=sx1+data['x1'][i]
            sy1=sy1+data['x2'][i]
        elif(arr[i]==2):
            sx2=sx2+data['x1'][i]
            sy2=sy2+data['x2'][i]
        else:
            sx3=sx3+data['x1'][i]
            sy3=sy3+data['x2'][i]
    nc_x.append(sx1/c1)
    nc_x.append(sx2/c2)
    nc_x.append(sx3/c3)
    print(nc_x)
    nc_y.append(sy1/c1)
    nc_y.append(sy2/c2)
    nc_y.append(sy3/c3)
    print(nc_y)
    for i in range(0,len(nc_x)):
        sub1 = nc_x[i]-c_x[i]
        sub2 = nc_y[i]-c_y[i]
    k+=1
```
